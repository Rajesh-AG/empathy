```python
import gradio as gr
from transformers import pipeline

# Load a lightweight emotion detection model
emotion_model = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")

# Function to analyze empathy level and respond
def chatbot_response(user_input, history):
    result = emotion_model(user_input)
    detected_emotion = result[0]['label'].lower()  # ensure lowercase for matching

    response_map = {
        "joy": "I'm glad to hear that! üòä",
        "sadness": "I'm here for you. Feel free to share more. üíô",
        "anger": "I understand you're upset. Take a deep breath. üòå",
        "fear": "That sounds scary. I'm here to listen. üíú",
        "love": "That's wonderful! Tell me more. ‚ù§Ô∏è",
        "surprise": "Oh! That must have been unexpected. ü§î",
        "confusion": "It's okay to be confused. Let's figure it out together. ü§ù",
        "disgust": "That's really unpleasant. Want to talk about it? üò£",
        "boredom": "Let‚Äôs find something interesting to do! üé≤"
    }

    response = response_map.get(detected_emotion, "I'm here for you. Tell me more.")

    return f"**Detected Emotion:** {detected_emotion.capitalize()}\n\n**Chatbot:** {response}"

# Gradio UI
demo = gr.Interface(fn=chatbot_response, inputs="text", outputs="text", title="ü§ñ Lightweight Empathy AI Bot")

# Launch the app
demo.launch(share=True)
```
